```python
# Declaring the Dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset

df = pd.read_csv("mnist_train.csv")
```

```python
# Display the first image
a = df.iloc[3, 1:].values  # Select the 4th row (index 3) and all columns except the first one
a = a[:784]  # Remove the last element (label) to get 784 pixel values
a = a.reshape(28, 28).astype('uint8')  # Reshape to 28x28 and convert to uint8

plt.imshow(a, cmap='gray')  # Use grayscale colormap for better visualization
plt.axis('off')  # Turn off axis for cleaner display
plt.show()
```

```python
# Split the dataset into features and labels

df_x = df.iloc[:, 1:]
df_y = df.iloc[:, 0]
```

```python
# Split the data into training and testing sets

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=4)
```

```python
# Initialize and train a random forest classifier

rf = RandomForestClassifier(n_estimators=100)
rf.fit(x_train, y_train)
```

```python
# Make predictions on the test set

pred = rf.predict(x_test)
```

```python
# Evaluate the accuracy of the model

correct_count = np.sum(pred == y_test.values)
accuracy = correct_count / len(pred)
```

```python
# Print the results

print("Predictions:", pred)
print("Correct Count:", correct_count)
print("Total Samples:", len(pred))
print(f"Accuracy: {accuracy * 100:.2f}%")
```
